To load the data as pandas dataframe 2 functions are needed. Those are fortunately found on the source page.  

```python
#import libraries
import gzip
import json

#functions copied from source
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

#use functions to create dataframe
df = getDF('data/raw/Toys_and_Games_5.json.gz')
```

### Data cleaning
some columns are not of the type indicated in the source. We are going to modify this.  
  
Vote is of type `object`, according to the dataset page this really should be a numeric type. After looking into it I found out some values are writen as `1,000` in stead of `1000` So lets remove all the comma's `,` and make it numeric.  
  
Additionally, currently there are two columns that refer to the time when the review was posted. We are going to replace these with a column of type datetime. 

```python
#clean out ','
df['vote'] = df['vote'].str.replace(",", "")
#change to numeric type
df['vote'] = pd.to_numeric(df['vote'])

#change to datetime type
df['unixReviewTime'] = pd.to_datetime(df['unixReviewTime'],unit='s')
#remove old review time column
del df['reviewTime']
#rename column so 'unix' is removed from column name
df = df.rename(columns = {'unixReviewTime': 'reviewTime'}, inplace = False)

df.info()
```

### Add a Sentiment colum

To make classification easier later on, it is useful and convenient to add a new column. This will be called `sentiment`, and has the following values: 

if `overall` Is `3`        the sentiment is `NEUTRAL`  
if `overall` Is `4` or `5` the sentiment is `POSITIVE`  
if `overall` Is `1` or `2` the sentiment is `NEGATIVE`


```python
def getSentiment(overall):
    sentiment = 'NEUTRAL'
    if overall > 3:
        sentiment = 'POSITIVE'
    if overall < 3:
        sentiment = 'NEGATIVE'
    return sentiment

df['sentiment'] = df['overall'].apply(getSentiment)

df.tail()
```

How is Sentiment distributed?

```python
fig, ax = plt.subplots()
df['sentiment'].value_counts().plot(ax=ax, kind='pie', autopct='%1.1f%%', startangle=90)
plt.show()
```

This pie graph really shows that the sentiment colum is not evenly distributed.

### Data filtering
For this project, we are only interested in two things:

1. A piece of text
2. The sentiment

So let's filter out all other data.

Other datasets have no `NEUTRAL` values. So we are going to filter these out as well.

```python
#filter on text and sentiment
df = df.filter(['reviewText', 'sentiment'], axis=1)

#filter out NEUTAL
df = df[df['sentiment'] != 'NEUTRAL']

#shuffle
df = df.sample(frac=1, random_state=42)
#drop index
df = df.reset_index(drop=True)

df
```
