## Data cleaning 
I found out that the word count provided by the dataframe is not correct.

```python
#positive
print("pos sentence: ", df['Positive_Review'][5])
print("dataframe word count : ", df['Review_Total_Positive_Word_Counts'][5])
print("calculated word count: ", len(df['Positive_Review'][5].split()))

#negative
print("neg sentence: ", df['Negative_Review'][5])
print("dataframe word count : ", df['Review_Total_Negative_Word_Counts'][5])
print("calculated word count: ", len(df['Negative_Review'][5].split()))
```

So I created a function that overrides the wordcount 

```python
def fixWordCount():
    df['Review_Total_Positive_Word_Counts'] = df['Positive_Review'].str.split().apply(len)
    df['Review_Total_Negative_Word_Counts'] = df['Negative_Review'].str.split().apply(len)

fixWordCount()
```

`Review_Date` is not of type datetime. Let's fix that

```python
df['Review_Date'] = pd.to_datetime(df['Review_Date'])
```

In each row there is a field for positive and negative text. While filling out, many people have entered text such as "absolutely **nothing**" or "__nothing__ at all" in the input field. We want to filter this out of the dataset, because "**nothing** at all" is not a good training text. We achieve this by filtering for the `keyword` **nothing** and then replacing the entire text with an empty string. But we only want to do this above a certain `word cap` because phrases like "the staff was very bad they did **nothing** at all and the food was very bad." do make good training sentences. 
  
Furthermore, when people have not filled in anything it is replaced with **No Negative** or **No Positive**. We also want to replace this text with empty strings. Lets add those words to the `keywords`.

```python
keywords = ['nothing' , 'Nothing', 'No Negative', 'No Positive']
wordcap = 15

df.loc[(df['Negative_Review'].str.contains('|'.join(keywords))) & (df['Review_Total_Negative_Word_Counts'] < wordcap), 'Negative_Review' ] = ""
df.loc[(df['Positive_Review'].str.contains('|'.join(keywords))) & (df['Review_Total_Positive_Word_Counts'] < wordcap), 'Positive_Review' ] = ""
```

Now we need to rerun the word count fixer because we deleted some reviews

```python
fixWordCount()
```

## filter data
We filter the data or the following way:
1. use only reviews with a word count above 0
2. split the data into two dataframes (one for positive one for negative)
3. add a sentiment column
4. rename the text column (so it matches other dataframes)

```python
# get reviews that have positive text
df_pos = df[df['Review_Total_Positive_Word_Counts'] > 0]
#get text data
df_pos = df_pos.filter(['Positive_Review'], axis=1)
#add sentiment
df_pos['sentiment'] = "POSITIVE"
#rename text colum to match amazon dataframe
df_pos = df_pos.rename(columns = {'Positive_Review': 'reviewText'}, inplace = False)
display(df_pos)


df_neg = df[df['Review_Total_Negative_Word_Counts'] > 0]
#get text data
df_neg = df_neg.filter(['Negative_Review'], axis=1)
#add sentiment
df_neg['sentiment'] = "NEGATIVE"
#rename text colum to match amazon dataframe
df_neg = df_neg.rename(columns = {'Negative_Review': 'reviewText'}, inplace = False)
display(df_neg)
```

Now we need to merge the two separate data frames back together again

```python
#combine
df_combined = pd.concat([df_pos, df_neg])
#shuffle
df_combined = df_combined.sample(frac=1, random_state=42)
#drop index
df_combined = df_combined.reset_index(drop=True)
df_combined
```
