Let's make sure each source has an equal number of positive and negative reviews and then put them back together. In this way, the data on source and sentiment is simultaneously distributed as fairly as possible. 

```python
#create equalizer function
def sentiment_equalizer(df):
    #split by sentiment
    df_pos = df.loc[df['sentiment'] == 'POSITIVE']
    df_neg = df.loc[df['sentiment'] == 'NEGATIVE']
    
    #check length
    if len(df_pos) > len(df_neg):
        #equalise
        df_pos = df_pos[:len(df_neg)]
    else:
        #equalise
        df_neg = df_neg[:len(df_pos)]
    
    #combine
    return pd.concat([df_neg, df_pos])
```

```python
#apply equalizer function for each source
equal_hotel = sentiment_equalizer(df_combined[df_combined['source'] == 'HOTEL'])
equal_amazon = sentiment_equalizer(df_combined[df_combined['source'] == 'AMAZON'])
equal_movie = sentiment_equalizer(df_combined[df_combined['source'] == 'MOVIE'])

#combine
df_combined = pd.concat([equal_hotel, equal_amazon, equal_movie])
#shuffle
df_combined = df_combined.sample(frac=1, random_state=42)
#drop index
df_combined = df_combined.reset_index(drop=True)
```

# Working with text data
According to sklearn's site on the page [working with text data](https://scikit-learn.org/stable/tutorial/text_analytics/working_with_text_data.html#extracting-features-from-text-files). Is bag of words the easiest way to recognize parts from text. To create a bag of words, CountVectorizer is used in the example. 

## Create bag of words

```python
# import CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

Bag of words works in the following way: The input is a list of sentences. The output is a matrix with all the words that appear in these sentences. And for each sentence, a matrix with indexes referring to the words in the sentences. 

![bag of words](img/bagOfWord.png)
This image explains it well. At the top you see a sentence. In the middle you see the matrix with indexes. And below the back of words. 

```python
# these are the input sentences
input_sentences = [
    "Here you have a test sentence",
    "This is another test sentence",
    "The word sentence is double in this sentence"
]

#create vectorizer
vect = CountVectorizer()

#create the matrix
matrix = vect.fit_transform(input_sentences)

#get the bag of words
bag_of_words = vect.get_feature_names()

#display
display(bag_of_words)
display(input_sentences)
display(matrix.toarray())
```

To get a nice overview of which words occur most often we are going to sum up the matrix. 

```python
matrix_total = np.asarray(matrix.sum(axis=0))[0]
matrix_total
```

We can use this summed matrix to create a nice graph that shows a good insight into word usage. 

```python
figure(figsize=(15, 10), dpi=80)

x = np.sort(matrix_total)
y = [x for _, x in sorted(zip(matrix_total,np.asarray(bag_of_words)))]

plt.barh(y, x)
plt.xticks(range(0,max(matrix_total)+1,1))
plt.grid(axis="x")
plt.show()
```

### Real data
Now that it is clear how bag of words works, we can apply it to the real review texts.

```python
%%time

#create vectorizer
vect = CountVectorizer()

#create the matrix
matrix = vect.fit_transform(df_combined['reviewText'])

#get the bag of words
bag_of_words = vect.get_feature_names()

#display
display(pd.Series(bag_of_words))
display(df_combined['reviewText'][:3])
display(matrix[:3].toarray())
```

We can see a nice beginning here. But we also see a problem here: `zzzzzzzzzzzzzzzzzzzzzz` is not really a useful word to train with
#### Filtering bag of words
To be more precise and create a more powerfull bag of words. I wrote a function that filters for numbers in words and checks if a word is in the dictionary. 

```python
#import library's
import enchant #dictionary library
import re #regex library

#create dictionary object
dictionary = enchant.Dict("en_US")

#create array to store filterd words
filterd_bag_of_words = []

for word in bag_of_words:
    #use regex to filer if word contains digets (=numeric)
    numeric = bool(re.search(r'\d', word))
    
    #use enchant to filter if word is in english dictionary
    is_a_word = dictionary.check(word)
    
    #if word is in dictionary and if nog numeric add to filterd array
    if (is_a_word and not(numeric)):
        filterd_bag_of_words.append(word)
        
#display filterd words
display(pd.Series(filterd_bag_of_words))
```

Now that we have created a `filterd_bag_of_words`. We need to create a new vectorizer with and fill it with the filtered words. 

```python
#Create new vectorizer
vect = CountVectorizer()
#fit with filterd words
vect.fit(filterd_bag_of_words)
```

```python
#check if words are indeed filterd
bag_of_words = vect.get_feature_names()
pd.Series(bag_of_words)
```

```python
# make matrix for all reviews with new bag of words
matrix = vect.transform(df_combined['reviewText'])
```

To make an even better comparison, we are going to compare the most frequently used words of a data frame with how often it appears in the other data frames. The source distribution is not equal. So we need to normalize the word count to make a fair distribution.

```python
#import library
from sklearn import preprocessing

# create normalise function
def normalise(arr): 
    normalised = preprocessing.normalize([arr])
    normalised = normalised[0]
    normalised = normalised / max(normalised) * 1
    return normalised
```

```python
# use normalise function to create normalised data
hotel_matrix_total_normalised = normalise(hotel_matrix_total)
amazon_matrix_total_normalised = normalise(amazon_matrix_total)
movie_matrix_total_normalised = normalise(movie_matrix_total)
```

Now that the normalized data is there. We can create a function that compares the most frequently used words, from a given dataset, to the other datasets. 

```python
#create function 
def plot_word_matrix_normalised(arr,arr_name, comp1, comp1_name, comp2, comp2_name):
    #Order labels by compared to arr
    labels = [x for _, x in sorted(zip(arr,np.asarray(bag_of_words)))][-50:]

    #Order arr, comp1 and comp2 counts by arr order
    arr_counts = [x for _, x in sorted(zip(arr,np.asarray(arr)))][-50:]
    comp1_counts = [x for _, x in sorted(zip(arr,np.asarray(comp1)))][-50:]
    comp2_counts = [x for _, x in sorted(zip(arr,np.asarray(comp2)))][-50:]

    x = np.arange(len(labels))  # the label locations
    width = 0.25  # the width of the bars

    fig, ax = plt.subplots()

    #size
    fig.set_figheight(15)
    fig.set_figwidth(15)

    #do hbar plots
    arr_bar   = ax.barh(x + width, arr_counts  , width, label=arr_name  , color= 'red')
    comp1_bar = ax.barh(x        , comp1_counts, width, label=comp1_name, color= 'blue')
    comp2_bar = ax.barh(x - width, comp2_counts, width, label=comp2_name, color= 'green')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_title('Word count of '+ arr_name + ' compared to: '+ comp1_name+ ' and '+comp2_name)
    ax.set_ylabel('Words')
    ax.set_yticklabels(labels)
    ax.set_xlabel('Count')
    ax.set_yticks(x)
    ax.legend()

    #add grid
    plt.xticks(np.arange(0, 1.1, 0.1))
    plt.grid(axis="x")

    fig.tight_layout()

    plt.show()
```

For each dataset, let's look only at the overrepresentation. Because underrepresentation is not visible in these graphs. This only becomes visible when you start comparing from another dataset.

### Difference in occurrences
You can clearly see here that there are clear differences in occurrences. To draw any real conclusions we need to focus on the number of differences in occurrences. So let's make a neat graph focussing on the differences.

```python
# function that calculates the difference 
def diff_calculator(pos, neg):
    diff_arr = []
    for ind in range(len((pos))):
        diff = pos[ind] - neg[ind]
        diff_arr.append(diff)
        
        #print first 5 elements to demonstrate what de function does.
        if ind < 5:
            print('pos:', pos[ind], 'neg:', neg[ind], 'diff:', diff)
    return diff_arr

#use function
diff_pos_neg = diff_calculator(pos_matrix,neg_matrix)
```

```python
#Sort lables by absolute (positive) diff values
diff_labels = [x for _, x in sorted(zip(map(abs,diff_pos_neg),np.asarray(bag_of_words)))]
#Sort diff values by absolute (positive) diff values
diff = [x for _, x in sorted(zip(map(abs,diff_pos_neg),np.asarray(diff_pos_neg)))]
```

```python
#make figure bigger
figure(figsize=(15, 10), dpi=80)

fig, ax = plt.subplots()

#size
fig.set_figheight(12)
fig.set_figwidth(15)

ax.set_title('Most used words seperated by sentiment')

positive = pd.Series(diff)[-50:] > 0

#use top 50 words to make bar diagram
ax.barh(diff_labels[-50:], diff[-50:], color=positive.map({True: 'g', False: 'r'}) )
#add x grid 
ax.grid(axis="x")

#add labes on x = 0 for readability
for ind in range(50):
    if np.array(positive)[ind]:
        ha = 'left'
    else:
        ha = 'right'
    ax.text(0, -.5+ind, diff_labels[-50:][ind],
            ha=ha, va='bottom')

#add label that shows negative vs positive
plt.xlabel("<-- NEGATIVE | POSITIVE -->")
plt.show()
```

### Word use explanation 
In this chart you can clearly see the difference in word usage. The word "the" was used almost 300,000 times more often in negative reviews than in positive reviews. This example is probably coincidental. But there are also words that stand out:

- **great, good, nice, excellent, perfect, lovely**
    - These are all words that belong to liking something. So it is also natural that these words were used more in **positive** reviews. 
- **location, staff**
    - These words probably come from the hotel review dataset. However, it is noticeable that these components are more often reviewed positively than negatively. 
- **friendly, helpful, clean, comfortable**
    - These are also likely to be words from the hotel review dataset. So apparently reviewers think it's very important that things are clean and comfortable. And people are friendly and helpful. 
  
  
- **and, but**
    - These words are both linking words. They do mean something different. **And** is a word that indicates **addition**, while **but** signals **contrast**. So we can assume that **positive** text **lists** things more often. And **negative** reviews more often use a **contradiction**. 

- **not**
    - not is a word often used in negative reviews. It is a word used in opposites. This is another confirmation that **contradictions** are used a lot in **negative** reviews. 
- **was, had**
    - These words are both written in past tense. It is likely that **negative** text was written in **past tense** more often than not.
